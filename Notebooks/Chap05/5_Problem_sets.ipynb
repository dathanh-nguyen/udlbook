```python
import numpy as np
import matplotlib.pyplot as plt
```

# Problem 5.2

```python
def cross_entropy_loss(sigmoids, y):
    loss = (-(1-y)*np.log(1-sigmoids)) - (y*np.log(sigmoids))
    return loss
```

```python
example_sigmoids = np.arange(0.01,1,0.01)
```

```python
fig, ax = plt.subplots()

ax.plot(example_sigmoids, cross_entropy_loss(example_sigmoids, 1), color = 'red', label = 'y=1')

ax.plot(example_sigmoids, cross_entropy_loss(example_sigmoids, 0), color = 'blue', label = 'y=0')

ax.set_xlabel('Sigmoid Values')

ax.set_ylabel('Cross Entropy Loss')

ax.set_title('Binary Cross-Entropy for y=0 and y=1')

ax.grid(True)
ax.legend()

plt.show()
```

We can see that at sigmoid values close to the corresponding y-value, the loss is lower, when the sigmoid value is opposite to the corresponding y-value, then it is greater.
